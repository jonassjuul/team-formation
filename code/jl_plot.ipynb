```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def import_res(filename,res) :
    f = open(filename,'r')
    f.readline()
    for line in f:
        line.strip()
        columns = line.split(',')
        for column in columns :
            column_split = column.split(':')
            if (column_split[0] in res.keys()) :
                #print(res)
                res[column_split[0]].append(float(column_split[1]))
            else : 
                if ("p" in res.keys()) :
                    res[column_split[0]] = [0]*len(res["p"])
                    res[column_split[0]].append(float(column_split[1]))

                else : 
                    res[column_split[0]] = [float(column_split[1])]

        for key in res.keys() :
            if (len(res[key])<len(res["p"])) :
                res[key].append(0)    
    
    return res
```

```python
res = {}
```

```python
res[3] = {}

filename_3 = "Outputs/jl_3simplex_N100_m3.txt"         
res[3] = import_res(filename_3,res[3])        
        
    
```

```python
plt.figure()
for key in res[3] :
    if (key != "p"):
        plt.plot(res[3]["p"],res[3][key],'k')
    
plt.xscale('log')
```

```python
res[4] = {}
filename_4 = "Outputs/jl_4simplex_N50_m4_Nexp5.txt"

res[4] =import_res(filename_4,res[4])
```

```python
emphasize=['3^1','3^4','3^3','3^2']
plt.figure()
for key in res[4] :
    if (key != "p"):
        if (key not in emphasize) :
            plt.plot(res[4]["p"],res[4][key],'k',alpha=0.5)
        else : 
            plt.plot(res[4]["p"],res[4][key],label=r'$%s$'%key)
    
plt.xscale('log')
plt.legend(loc=1,frameon=False)
```
